### Loading Libraries

```
import matplotlib.pyplot as plt #plotting
from datetime import datetime, timedelta #data creation 
from celluloid import Camera #animation
plt.style.use('seaborn')
```

### Create sample data for animation

```


dates = [
    datetime(2020, 8, 11),
    datetime(2020, 8, 12),
    datetime(2020, 8, 13),
    datetime(2020, 8, 14),
    datetime(2020, 8, 15),
    datetime(2020, 8, 16),
    datetime(2020, 8, 17),
    datetime(2020, 8, 18),
    datetime(2020, 8, 19),
    datetime(2020, 8, 20),
    datetime(2020, 8, 21),
    datetime(2020, 8, 22),
    datetime(2020, 8, 23),
    datetime(2020, 8, 24)
]

y = [0, 4, 3, 1, 3, 2, 6,0, 4, 3, 1, 3, 2, 6]

```

### 1st step in Animation - Fig + Camera

```

fig = plt.figure()
camera = Camera(fig)
```

```
plt.plot_date(dates,y, linestyle = 'solid')
```

```
for i in range(1,len(y)+1):
    new_dates = dates[:i]
    z = y[:i]
    plt.plot_date(new_dates,z, linestyle = 'solid')
    plt.tight_layout()
    plt.show()
    camera.snap()
```

```
animation = camera.animate()
```

```
import numpy as np
from matplotlib import pyplot as plt
from celluloid import Camera

fig, axes = plt.subplots(2)
camera = Camera(fig)
t = np.linspace(0, 2 * np.pi, 128, endpoint=False)
for i in t:
    axes[0].plot(t, np.sin(t + i), color='blue')
    axes[1].plot(t, np.sin(t - i), color='blue')
    camera.snap()
animation = camera.animate()
```
